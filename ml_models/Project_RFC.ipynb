```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/ml_proccessed_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

Converting object datatypes to integer

```python
df['Group'] = df['Group'].replace(['Demented', 'Converted', 'Nondemented'], [0, 1, 2])
```

```python
df.head()
```

```python
df.info()
```

```python
df['M/F'] = df['M/F'].replace(['M', 'F'], [0, 1])
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
# saving the dataframe
df.to_csv('ml_model.csv')
```

Separating dependent and independent variables

```python
y = df['Group']
X = df.drop('Group', axis=1)
```

Creating training and testing datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

Training Model using random forest classification


```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
predictions = rf.predict(X_test)
```

Metrics calculation of random forest classification

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.metrics import mean_squared_error
rfrmodel_rmse = np.sqrt(mean_squared_error(predictions, y_test))
print("RMSE value for Random Forest is:",rfrmodel_rmse)
```

Training Model using support vector machine

```python
from sklearn.svm import SVC
svm_model = SVC()
svm_model = svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
```

Metrics calculation of support vector machine

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.metrics import mean_squared_error
svmmodel_rmse = np.sqrt(mean_squared_error(y_pred, y_test))
print("RMSE value for Support Vector Machine is:",svmmodel_rmse)
```

**Comparing both the algorithms, random forest classification and support vector machine, it is clearly seen that random forest classificarion is the best algorithm with greater accuracy and best RMSE score**
